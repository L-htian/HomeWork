```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_csv("over_point_to_case_type.csv",index_col=[0])
data1
```

```python
data2 = pd.read_csv("over_point_to_user.csv",index_col=[0])
data2
```

```python
day_count = data2["days"].value_counts().drop(labels=[-1,-2])
day_count
```

```python
day_count.sort_index(inplace=True)
day_count
```

```python
plt.plot(day_count)
plt.xlabel("days")
plt.ylabel("number_of_submissions")
plt.savefig("image/number_of_submissions_by_days")
```

```python
data4 = data2.drop_duplicates(subset=["case_id","user_id"])
data4
```

```python
day_count2 = data4["days"].value_counts().drop(labels=[-1,-2])
day_count2
```

```python
day_count2.sort_index(inplace=True)
day_count2
```

```python
plt.plot(day_count2)
plt.xlabel("days")
plt.ylabel("number_of_last_submissions")
plt.savefig("image/number_of_last_submissions_by_days")
```
