```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_csv("over_point_to_case_type.csv",index_col=[0])
data1
```

```python
data2 = pd.read_csv("over_point_to_user.csv",index_col=[0])
data2
```

```python
case_type = []
for i in range(87734):
    a = data1.iloc[i].loc["case_type"]
    if a == "字符串":
        case_type.append("string")
    if a == "数组" :
        case_type.append("array")
    if a== "线性表" :
        case_type.append("linear_table")
    if a== "图结构" :
        case_type.append("graph_structure")
    if a== "查找算法" :
        case_type.append("search_algorithm")
    if a== "树结构" :
        case_type.append("tree_structure")
    if a== "排序算法" :
        case_type.append("sorting_algorithm")
    if a== "数字操作" :
        case_type.append("digital_operation")
case_type
```

```python
data1["case_type"] = case_type
data2["case_type"] = case_type
```
