```python
import pandas as pd
import json
```

```python
raw_data = pd.read_json("test_data.json")
raw_data
```

```python
result1 = pd.DataFrame(columns=("case_id","case_type","case_zip","final_score","user_id"))
result1
```

```python
for i in range(len(raw_data.columns)):
    case = (raw_data.iloc[1].iloc[i])
    for k in range(len(case)):
        temp = pd.DataFrame(case[k])
        temp["user_id"] = raw_data.iloc[0].iloc[i]
        result1 = pd.concat([result1,temp])
result1
```

```python
result1.index = range(len(result1))
result1
```

```python
result1["index"] = result1.index
result1
```

```python
del result1["case_zip"]
result1
```

```python
uRecords = result1.loc[:,"upload_records"]
uRecords
```

```python
result2 =pd.DataFrame(columns=("upload_id","upload_time","code_url","score","index"))
result2
```

```python
for i in range(len(uRecords)):
    uRecord = pd.DataFrame(uRecords[i],index=[0])
    uRecord["index"] = result1.loc[:,"index"].iloc[i]
    result2 = pd.concat([result2,uRecord])
result2
```

```python
del result2["upload_id"]
del result2["code_url"]
result2
```

```python
del result1["upload_records"]
result1
```

```python
result = pd.merge(result1,result2,on="index")
result
```

```python
del result["index"]
result
```

```python
result.to_csv("data.csv")
result.to_json("data.json")
```
