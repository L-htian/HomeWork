```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_csv("over_point_to_case_type.csv",index_col=[0])
data1
```

```python
timestamps = []
for i in range(87734):
    timestamp = data1.iloc[i].loc["upload_time"] + 28800000
    timestamps.append(timestamp)
data1["upload_time"] = timestamps
```

```python
data2 = data1
data2["upload_time"] = pd.to_datetime(data2["upload_time"],unit="ms")
data2
```

```python
days = []
for i in range(87734):
    month = data2.iloc[i].loc["upload_time"].strftime("%m")
    month = int(month)
    day = data2.iloc[i].loc["upload_time"].strftime("%d")
    day = int(day)
    day = day - 16 + (month - 2) * 29
    days.append(day)
data2["days"] = days
```

```python
data2
```

```python
hours = []
for i in range(87734):
    hour = data2.iloc[i].loc["upload_time"].strftime("%H")
    hour = int(hour)
    minute = data2.iloc[i].loc["upload_time"].strftime("%M")
    minute = int(minute)
    hour = hour + ((int)(minute/12))/5.0
    hours.append(hour)
data2["hours"] = hours
data2
```

```python
data2["upload_time_stamp"] = data1["upload_time"]
```
