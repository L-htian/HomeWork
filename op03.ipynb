```python
import pandas as pd
import json 
```

```python
final_scores = pd.read_csv("final_scores.csv")
final_scores
```

```python
upload_times = pd.read_csv("upload_times.csv")
upload_times
```

```python
del upload_times["Unnamed: 0"]
del final_scores["Unnamed: 0"]
```

```python
average_score_of_everyone = final_scores
```

```python
sum_score = average_score_of_everyone.groupby("user_id").sum()
sum_score
```

```python
sum_score = sum_score.sort_values("over_point")
```

```python
sum_score.index = range(len(sum_score))
sum_score
```

```python
import matplotlib.pyplot as plt
plt.plot(sum_score.loc[:,"over_point"])
```
