```python
data4 = pd.read_csv("data.csv",index_col=[0])
data4
```

```python
data4["upload_time"] = pd.to_datetime(data4["upload_time"],unit="ms")
hours = []
for i in range(87734):
    hour = int(data4.iloc[i].loc["upload_time"].strftime("%H"))
    minute = int(data4.iloc[i].loc["upload_time"].strftime("%M"))
    hour = minute/60.0 + hour
    hours.append(hour)
data4["minutes"] = hours
data4
```

```python
data4["hours"] = hours
plt.figure(figsize=(15, 10))
sns.regplot(x = 'hours',y = 'score',data = data4, scatter_kws={'s':1}, fit_reg=False)
plt.show()
```

```python
data4 = data4.drop_duplicates(subset=["case_id","user_id"],keep="last")
data4
```

```python
plt.figure(figsize=(15, 10))
sns.regplot(x = 'minutes',y = 'score',data = data4, scatter_kws={'s':1}, fit_reg=False)
plt.show()
```
