```python
data6 = data4.drop_duplicates(subset=["case_id","user_id"])
data6
```

```python
data7 = data6.groupby("case_type")
```

```python
fig2 = plt.figure(figsize=(15,10))
ax2 = fig2.add_subplot(1,1,1)
```

```python
array = data7.get_group("array")
array_day_count = array["days"].value_counts()
array_day_count.sort_index(inplace=True)
ax2.plot(array_day_count,"b",label="array")

digital_operation = data7.get_group("digital_operation")
digital_operation_day_count = digital_operation["days"].value_counts()
digital_operation_day_count.sort_index(inplace=True)
ax2.plot(digital_operation_day_count,"g",label="digital_operation")

graph_structure = data7.get_group("graph_structure")
graph_structure_day_count = graph_structure["days"].value_counts()
graph_structure_day_count.sort_index(inplace=True)
ax2.plot(graph_structure_day_count,"r",label="graph_structure")

linear_table = data7.get_group("linear_table")
linear_table_day_count = linear_table["days"].value_counts()
linear_table_day_count.sort_index(inplace=True)
ax2.plot(linear_table_day_count,"c",label="linear_table")

search_algorithm = data7.get_group("search_algorithm")
search_algorithm_day_count = search_algorithm["days"].value_counts()
search_algorithm_day_count.sort_index(inplace=True)
ax2.plot(search_algorithm_day_count,"m",label="search_algorithm")

sorting_algorithm = data7.get_group("sorting_algorithm")
sorting_algorithm_day_count = sorting_algorithm["days"].value_counts()
sorting_algorithm_day_count.sort_index(inplace=True)
ax2.plot(sorting_algorithm_day_count,"y",label="sorting_algorithm")

string = data7.get_group("string")
string_day_count = string["days"].value_counts()
string_day_count.sort_index(inplace=True)
ax2.plot(string_day_count,"k",label="string")

tree_structure = data7.get_group("tree_structure")
tree_structure_day_count = tree_structure["days"].value_counts()
tree_structure_day_count.sort_index(inplace=True)
ax2.plot(tree_structure_day_count,"silver",label="tree_structure")

ax2.legend(loc = "best")
ax2.set_xlabel("hours")
ax2.set_ylabel("num_of_submissions")
fig2.savefig("image/number_of_submissions_by_hours(group_by_casetype)")
```
